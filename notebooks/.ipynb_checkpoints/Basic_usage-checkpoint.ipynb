# Basic usage

```python
import DGS
```

### Set parameters

```python
density = 10 # process every 10 lines
resolution = 1 # mm/pixel, i.e. no scale
dofilter =1 # filter the image
notes = 16 # notes per octave
maxscale = 8 #Max scale as inverse fraction of data length
verbose = 1 # print stuff to screen
x = -0.5 #no area to volume conversion
```

### Select an image to process

```python
image_file = '../DGS/IMG_0229.JPG'
```

### Run DGS

```python
dgs_stats = DGS.dgs(image_file, density, resolution, dofilter, maxscale, notes, verbose, x)
```

### Plot the distribution and percentiles

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.semilogx(dgs_stats['grain size bins'], dgs_stats['grain size frequencies'], 'k-o', label='GSD')
plt.legend()
plt.xlabel('Grain Size [pixels]')
plt.ylabel('Normalized frequency [-]')
```

```python
import numpy as np
```

```python
plt.figure(figsize=(10,10))
plt.plot(dgs_stats['grain size bins'][:5], np.cumsum(dgs_stats['grain size frequencies'][:5]), 'k-o', label='GSD')
plt.legend()
plt.xlabel('Grain Size [pixels]')
plt.ylabel('Normalized frequency [-]')
counter = 0
for k in dgs_stats['percentile_values']:
    plt.axvline(k)
    plt.text(k,1.03,str(dgs_stats['percentiles'][counter]), rotation=45)
    counter += 1
```

```python
print('Percentiles: '+str(dgs_stats['percentile_values']))
print('Percentile values: '+str(dgs_stats['percentiles']))
print('Mean: '+str(dgs_stats['mean grain size']))
print('Arithmetic Sorting: '+str(dgs_stats['grain size sorting']))
print('Skewness: '+str(dgs_stats['grain size skewness']))
print('Kurtosis: '+str(dgs_stats['grain size kurtosis']))
```

### Batch process images

```python
from glob import glob

files = glob('../DGS/*.JPG')
files
```

```python
allstats = {}
for image_file in files:
    dgs_stats = DGS.dgs(image_file, density, resolution, dofilter, maxscale, notes, verbose, x)
    allstats[image_file] = dgs_stats
```

```python
from scipy.misc import imread
```

```python
plt.figure(figsize=(15,15))
counter = 1
for image_file in files:
    plt.subplot(2,3,counter)
    plt.imshow(imread(image_file)); plt.axis('off')
    plt.subplot(2,3,counter+3)
    plt.plot(allstats[image_file]['grain size bins'], allstats[image_file]['grain size frequencies'], label=image_file)
    plt.axvline(allstats[image_file]['mean grain size'], color='r')
    plt.xlim(0,400)
    if counter==1:
        plt.xlabel('Grain Size [pixels]')
        plt.ylabel('Normalized frequency [-]')
    counter += 1
```
